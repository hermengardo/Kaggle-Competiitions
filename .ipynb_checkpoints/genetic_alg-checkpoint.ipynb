```python
from kaggle_environments import make
from kaggle_environments.envs.orbit_wars.orbit_wars import Planet, Fleet, CENTER, ROTATION_RADIUS_LIMIT

import math
import numpy as np
import random
from orbit_wars.core.viz.embed import render
from tqdm import tqdm
from enum import Enum

from multiprocessing import Pool
from dataclasses import dataclass

env = make("orbit_wars", debug=True)
```

```python
# Map
BOARD = 100.0
SUN_X = 50.0
SUN_Y = 50.0
SUN_R = 10.0

# Physics
MAX_SPEED = 6.0
SUN_SAFETY = 1.5
LAUNCH_CLEARANCE = 0.2
CONTESTED_NEUTRAL_MARGIN = 2
INTERCEPT_TOLERANCE = 1
AIM_ITERATIONS = 5
ROTATION_LIMIT = 50.0

# Features
CANDIDATE_FEATURES = 12

# Model
MIN_SHIPS = 10
MASS_PROB = 0.5
HIDDEN = 32
N_WEIGHTS = CANDIDATE_FEATURES * HIDDEN + HIDDEN + HIDDEN + 1

# Utils
ID, OWNER, X, Y, RADIUS, SHIPS, PROD = 0, 1, 2, 3, 4, 5, 6
```

```python
def candidate_features(src, dst, frac, ships_sent, total_planets, owned_count, player, fleets_leaving_dst, fleets_leaving_src):
    return np.array([
        frac,
        owned_count / total_planets,
        math.log1p(src.ships) / math.log1p(1000),
        math.log1p(dst.ships) / math.log1p(1000),
        dist(src.x, src.y, dst.x, dst.y) / 100.0,
        math.log1p(fleets_leaving_dst) / math.log1p(1000),
        math.log1p(fleets_leaving_src) / math.log1p(1000),
        dst.radius,
        src.radius,
        float(dst.owner != player and dst.owner != -1),
        float(dst.owner == -1),
        math.log1p(ships_sent) / math.log1p(1000),
    ], dtype=np.float32)

def dist(ax, ay, bx, by):
    return math.hypot(ax - bx, ay - by)


def orbital_radius(planet):
    return dist(planet.x, planet.y, SUN_X, SUN_Y)


def is_static_planet(planet):
    return orbital_radius(planet) + planet.radius >= ROTATION_LIMIT


def fleet_speed(ships):
    if ships <= 1:
        return 1.0
    ratio = math.log(ships) / math.log(1000.0)
    ratio = max(0.0, min(1.0, ratio))
    return 1.0 + (MAX_SPEED - 1.0) * (ratio ** 1.5)


def point_to_segment_distance(px, py, x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    seg_len_sq = dx * dx + dy * dy
    if seg_len_sq <= 1e-9:
        return dist(px, py, x1, y1)
    t = ((px - x1) * dx + (py - y1) * dy) / seg_len_sq
    t = max(0.0, min(1.0, t))
    proj_x = x1 + t * dx
    proj_y = y1 + t * dy
    return dist(px, py, proj_x, proj_y)

def segment_hits_sun(x1, y1, x2, y2, safety=SUN_SAFETY):
    return point_to_segment_distance(SUN_X, SUN_Y, x1, y1, x2, y2) < SUN_R + safety


def launch_point(sx, sy, sr, angle):
    clearance = sr + LAUNCH_CLEARANCE
    return sx + math.cos(angle) * clearance, sy + math.sin(angle) * clearance


def actual_path_geometry(sx, sy, sr, tx, ty, tr):
    angle = math.atan2(ty - sy, tx - sx)
    start_x, start_y = launch_point(sx, sy, sr, angle)
    hit_distance = max(0.0, dist(sx, sy, tx, ty) - (sr + LAUNCH_CLEARANCE) - tr)
    end_x = start_x + math.cos(angle) * hit_distance
    end_y = start_y + math.sin(angle) * hit_distance
    return angle, start_x, start_y, end_x, end_y, hit_distance


def safe_angle_and_distance(sx, sy, sr, tx, ty, tr):
    angle, start_x, start_y, end_x, end_y, hit_distance = actual_path_geometry(
        sx, sy, sr, tx, ty, tr,
    )
    if segment_hits_sun(start_x, start_y, end_x, end_y):
        return None
    return angle, hit_distance

def predict_planet_position(planet, initial_by_id, angular_velocity, turns):
    init = initial_by_id.get(planet.id)
    if init is None:
        return planet.x, planet.y
    r = dist(init[2], init[3], SUN_X, SUN_Y)
    if r + init[4] >= ROTATION_LIMIT:
        return planet.x, planet.y
    cur_ang = math.atan2(planet.y - SUN_Y, planet.x - SUN_X)
    new_ang = cur_ang + angular_velocity * turns
    return (
        SUN_X + r * math.cos(new_ang),
        SUN_Y + r * math.sin(new_ang),
    )
```

```python
class Agent:
    def __init__(self, hidden=HIDDEN):
        self.h1 = hidden
        self.h2 = hidden // 2
        self.h3 = hidden // 2

        total_params = (
            CANDIDATE_FEATURES * self.h1 + self.h1 +
            self.h1 * self.h2 + self.h2 +
            self.h2 * self.h3 + self.h3 +
            self.h3 + 1
        )

        self.w = np.random.randn(total_params).astype(np.float32) * 0.1

    def _unpack(self):
        idx = 0

        W1 = self.w[idx:idx + CANDIDATE_FEATURES * self.h1].reshape(CANDIDATE_FEATURES, self.h1)
        idx += CANDIDATE_FEATURES * self.h1
        b1 = self.w[idx:idx + self.h1]
        idx += self.h1

        W2 = self.w[idx:idx + self.h1 * self.h2].reshape(self.h1, self.h2)
        idx += self.h1 * self.h2
        b2 = self.w[idx:idx + self.h2]
        idx += self.h2

        W3 = self.w[idx:idx + self.h2 * self.h3].reshape(self.h2, self.h3)
        idx += self.h2 * self.h3
        b3 = self.w[idx:idx + self.h3]
        idx += self.h3

        W4 = self.w[idx:idx + self.h3].reshape(self.h3, 1)
        idx += self.h3
        b4 = self.w[idx:idx + 1]

        return W1, b1, W2, b2, W3, b3, W4, b4

    def act(self, obs):
        player = obs.get("player", 0)
        planets = [Planet(*p) for p in obs.get("planets", [])]
        fleets = [Fleet(*f) for f in obs.get("fleets", [])]
        angular_vel = obs.get("angular_velocity", 0)
        initial_planets = obs.get("initial_planets", [])
        initial_by_id = {p[0]: p for p in initial_planets}
        comets = obs.get("comets", []) or []
        comet_ids = set(obs.get("comet_planet_ids", []) or [])

        owned = [p for p in planets if p.owner == player and p.ships > MIN_SHIPS]
        if not owned:
            return []

        owned_count = len(owned)
        total_planets = len(planets)
        fracs = [0.0, 0.20, 0.40, 0.60, 0.80, 1.00]

        candidates = []
        feat_matrix = []

        for src in owned:
            for dst in planets:
                if dst.id == src.id:
                    continue
                    
                if dst.id in comet_ids:
                    continue

                fleets_leaving_dst = sum(f.ships for f in fleets if f.from_planet_id == dst.id)
                fleets_leaving_src = sum(f.ships for f in fleets if f.from_planet_id == src.id)
    
                for frac in fracs:
                    ships_sent = max(0, int(frac * src.ships))
                    speed = fleet_speed(ships_sent)
                    travel_time = dist(src.x, src.y, dst.x, dst.y) / speed

                    if is_static_planet(dst):
                        dx, dy = dst.x, dst.y
                        travel_time = dist(src.x, src.y, dx, dy) / speed
                    else:
                        dx, dy = dst.x, dst.y
                        for _ in range(AIM_ITERATIONS):
                            travel_time = dist(src.x, src.y, dx, dy) / speed
                            dx, dy = predict_planet_position(dst, initial_by_id, angular_vel, travel_time)

                    safe = safe_angle_and_distance(src.x, src.y, src.radius, dx, dy, dst.radius)
                    if safe is None:
                        continue

                    angle = safe[0]

                    feats = candidate_features(src, dst, frac, ships_sent, total_planets, owned_count, player, fleets_leaving_dst, fleets_leaving_src)
                    candidates.append((src.id, angle, ships_sent))
                    feat_matrix.append(feats)

        if not candidates:
            return []

        W1, b1, W2, b2, W3, b3, W4, b4 = self._unpack()
        X = np.stack(feat_matrix)

        H1 = np.tanh(X @ W1 + b1)
        H2 = np.tanh(H1 @ W2 + b2)
        H3 = np.tanh(H2 @ W3 + b3)

        scores = (H3 @ W4 + b4).squeeze()
        scores = scores - scores.max()
        probs = np.exp(scores) / np.exp(scores).sum()
        order = np.argsort(probs)[::-1]

        actions = []
        used_sources = set()
        cumulative = 0.0
        for idx in order:
            src_id, angle, ships = candidates[idx]
            if src_id in used_sources:
                continue
            actions.append([src_id, angle, ships])
            used_sources.add(src_id)
            cumulative += probs[idx]
            if cumulative > MASS_PROB:
                break

        return actions
```

```python
def get_params(agent):
    return agent.w.copy()

def set_params(agent, params):
    agent.w = params.astype(np.float32)

def save(agent, path):
    np.save(path, get_params(agent))

def load(agent, path):
    set_params(agent, np.load(path))

def make_agent_fn(model):
    def fn(obs, config=None):
        return model.act(obs)
    return fn
```

```python
def crossover(father, mother):
    gf, gm = get_params(father), get_params(mother)
    mask = np.random.rand(len(gf)) > 0.5
    child = Agent()
    set_params(child, np.where(mask, gf, gm))
    return child

def mutate(agent, std=0.02):
    params = get_params(agent)
    params += np.random.randn(len(params)).astype(np.float32) * std
    set_params(agent, params)

def tournament(population, fitnesses, k=3):
    contestants = random.sample(list(zip(fitnesses[:len(population)], population)), k)
    winner = max(contestants, key=lambda x: x[0])
    return winner[1], winner[0]
```

```python
def compute_fitness(state, idx):
    reward = state[idx].reward or 0
    obs = state[idx].observation
    planets = [Planet(*p) for p in obs.planets]
    owned = [p for p in planets if p.owner == obs.player]
    
    high_owned = sum(1 for p in owned if p.production == 5)
    high_total = sum(1 for p in planets if p.production == 5)
    resource_wise = high_owned / high_total if high_total else 0
    domination = len(owned) / len(planets)
    
    resource_wise = resource_wise if reward >= 0 else 0
    domination = domination if reward >= 0 else 0

    return reward * 1000 + resource_wise * 800 + domination * 500
```

```python
def evaluate_pair(args):
    i, j, params_i, params_j = args

    agent_i = Agent()
    agent_j = Agent()
    set_params(agent_i, params_i)
    set_params(agent_j, params_j)

    fn_i = make_agent_fn(agent_i)
    fn_j = make_agent_fn(agent_j)

    players = [(fn_i, i), (fn_j, j)]
    random.shuffle(players)

    env = make("orbit_wars", debug=True)
    env.run([p[0] for p in players])

    idx_i = [p[1] for p in players].index(i)
    idx_j = [p[1] for p in players].index(j)

    fit_i = compute_fitness(env.state, idx_i) - 0.0001 * float(np.sum(params_i ** 2))
    fit_j = compute_fitness(env.state, idx_j) - 0.0001 * float(np.sum(params_j ** 2))

    return i, j, fit_i, fit_j
    
def evaluate_quad(args):
    indices, params_list = args
    agents = [Agent() for _ in range(4)]
    for agent, params in zip(agents, params_list):
        set_params(agent, params)

    agent_fns = [make_agent_fn(a) for a in agents]
    order = list(range(4))
    random.shuffle(order)

    env = make("orbit_wars", debug=True)
    env.run([agent_fns[o] for o in order])

    results = []

    for game_pos, original_pos in enumerate(order):
        fit = compute_fitness(env.state, game_pos) - 0.0001 * float(np.sum(params_list[original_pos] ** 2))
        results.append(
            (indices[original_pos], fit)
        )

    return results
```

```python
DECAY = .9995
```

```python
def sim(pop_size=20, n_gens=50, k=3, n_repeats=3, mutate_std=0.1, mutation_rate=0.1, workers=12):
    population = [Agent() for _ in range(pop_size)]
    best_global = None
    max_fit = float("-inf")
    history = {"best": [], "max": [], "mean": [], "gen": []}
    env = make("orbit_wars", debug=True)
    
    with Pool(workers) as pool:
        for gen in tqdm(range(n_gens), desc="Generations"):
            # Gera as chaves
            pairs = []
            for i in range(pop_size):
                pool_opponents = [j for j in range(pop_size) if j != i]
                pairs += [(i, j, get_params(population[i]), get_params(population[j])) for j in random.sample(pool_opponents, k=k)]
                    
            # Jogos
            all_results_1v1 = []
            for i in range(n_repeats):
                all_results_1v1.append(pool.map(evaluate_pair, pairs))
            
            
            fitnesses_1v1 = np.zeros(pop_size)
            counts_1v1 = np.zeros(pop_size)
            
            for results in all_results_1v1:
                for i, j, fit_i, fit_j in results:
                    fitnesses_1v1[i] += fit_i
                    fitnesses_1v1[j] += fit_j
                    counts_1v1[i] += 1
                    counts_1v1[j] += 1
                    
            fitnesses = np.mean(fitnesses_1v1) / np.maximum(counts_1v1, 1)
    
            mutate_std *= DECAY
            mutation_rate *= DECAY
            mutation_rate = max(mutation_rate, 0.05)
            mutate_std = max(mutate_std, 0.05)
            
            ranked = sorted(zip(fitnesses, population), key=lambda x: x[0], reverse=True)
            best_fit, best = ranked[0]
            
            if best_fit > max_fit - 250:
                if best_fit > max_fit:
                    max_fit = best_fit
                    best_global = best

                env = make("orbit_wars", debug=True)
                env.run([make_agent_fn(best), "starter"])
                with open(f"replay_agent_gen_{gen + 1}.html", "w", encoding="utf-8") as f:
                    f.write(env.render(mode="html"))
    
            save(best, f"best_agent_gen_{gen + 1}.npy")
            mean_fit = fitnesses.mean()
            tqdm.write(f"Gen {gen+1:3d} | best={best_fit:.1f} | mean={mean_fit:.1f} | max={max_fit:.1f} | std={mutate_std:.2f} | mutation rate={mutation_rate:.2f}")
            
            elite = [agent for _, agent in ranked[:3]]
            new_population = list(elite)
            while len(new_population) < pop_size:
                father, father_fit = tournament(population, fitnesses)
                mother, mother_fit = tournament(population, fitnesses)
                child = crossover(father, mother)
                
                if random.random() < mutation_rate:
                    mutate(child, std=mutate_std)
                
                new_population.append(child)


    save(best_global, "best_agent.npy")
    save(best, "last.npy")
    
    return history
```

```python
history = sim(pop_size=12, n_gens=10_000, k=1, n_repeats=3, mutation_rate=1, mutate_std=0.9)
```
