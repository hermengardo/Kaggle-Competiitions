This notebook demonstrates how to use the high-tech custom visualizer for the Orbit Wars environment. It includes a custom `render()` function that embeds a standalone HTML/JS visualization directly into the notebook.

### Features:
- 🚀 **Tactical HUD & Typography**
- 💥 **Combat Impact Effects**
- 📊 **APM Sparklines & Advantage Bars**
- 🔍 **Glassmorphism & CRT Monitor Aesthetics**

```python
from kaggle_environments import make
from orbit_wars.core.viz.embed import render

# Create the environment
env = make("orbit_wars", debug=True)

# Run a game (e.g. random vs random)
env.run(["random", "random"])

# Render the game using our custom visualizer!
render(env)
```
